```python
from transformers import *
import torch as t
import re
```

```python
config = Config()
model = Transformer(config)

if t.cuda.is_available():
    weights = t.load('./saved_runs/mod_digit_add_2024-09-13_15-58/450.pth')
else:
    weights = t.load('./saved_runs/mod_digit_add_2024-09-13_15-58/450.pth', map_location='cpu')
    
model.load_state_dict(weights['model'])
tokenizer = Tokenizer(config)
```

```python
test_sentence = "122+1="
test_sentence_tokenized = tokenizer.tokenize(test_sentence)
print(test_sentence_tokenized)

# pred = model.generate_greedy(test_sentence_tokenized)
# print("Test sentence:", test_sentence)
# print("Prediction:", tokenizer.detokenize(pred))
```

```python
# Case 1: Both operators are one digit

counter_correct_1 = 0
counter_total_1 = 0

for i in range(10):
    for j in range(10):
        counter_total_1 += 1
        test_sentence = f"{i}+{j}="
        result = (i+j)%113
        test_sentence_tokenized = tokenizer.tokenize(test_sentence)
        pred = model.generate_greedy(test_sentence_tokenized)
        pattern = re.compile(r'=')
        idx = pattern.search(test_sentence).start()
        res = pred[idx+1:-1]
        res_str = "".join(tokenizer.detokenize(res))
        int_pred = int(res_str)
        if int_pred == result:
            counter_correct_1 += 1
print(f"Accuracy for one digit operators: {counter_correct_1/counter_total_1}")

```
